```python

from sentence_transformers import SentenceTransformer
```

```python
# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")

```

```python
# Example of usage with daily sentences
# 
# The sentences to encode
sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium.",
]

# 2. Calculate embeddings by calling model.encode()
embeddings = model.encode(sentences)
print(embeddings.shape)
# [3, 384]

# 3. Calculate the embedding similarities
similarities = model.similarity(embeddings, embeddings)
print(similarities)
# tensor([[1.0000, 0.6660, 0.1046],
#         [0.6660, 1.0000, 0.1411],
#         [0.1046, 0.1411, 1.0000]])
```

```python
# Example of usage with Globecom conference names
# 
globecoms = ["GLOBECOM - IEEE Global Telecommunications Conference","Proceedings - IEEE Global Communications Conference, GLOBECOM",
             "GLOBECOM 2020 - IEEE Global Telecommunications Conference","Proceedings - IEEE Global Communications Conference, GLOBECOM 2018"]
embeddings = model.encode(globecoms)
print(embeddings.shape)
# [3, 384]

# 3. Calculate the embedding similarities
similarities = model.similarity(embeddings, embeddings)
print(similarities)

```
